#Célula 1: Importação das Bibliotecas





```python
# Importando bibliotecas necessárias
from sklearn.datasets import fetch_california_housing
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import RidgeCV
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_validate

# Importando bibliotecas necessárias
# - fetch_california_housing: Para carregar o dataset de habitação da Califórnia.
# - pandas: Para manipulação e análise de dados em formato tabular.
# - matplotlib e seaborn: Para criar visualizações gráficas dos dados.
# - numpy: Para operações numéricas e geração de números aleatórios.
# - StandardScaler: Para normalizar as features, garantindo que todas tenham a mesma escala.
# - RidgeCV: Para criar um modelo de regressão Ridge com validação cruzada.
# - make_pipeline: Para criar um pipeline que integra pré-processamento e modelagem.
# - cross_validate: Para avaliar o modelo usando validação cruzada.
```

#Célula 8: Scatter Plot da Localização Geográfica


```python
# Scatter plot da localização geográfica vs. valor das casas
sns.scatterplot(
    data=california_housing.frame,
    x="Longitude",
    y="Latitude",
    size="MedHouseVal",
    hue="MedHouseVal",
    palette="viridis",
    alpha=0.5,
)
plt.legend(title="MedHouseVal", bbox_to_anchor=(1.05, 0.95), loc="upper left")
plt.title("Median house value depending of their spatial location")
plt.show()

# Criando um scatter plot da localização geográfica vs. valor das casas
# - sns.scatterplot: Cria um gráfico de dispersão onde o eixo x é a longitude, o eixo y é a latitude, e o tamanho/cor dos pontos representa o valor das casas.
# - Isso ajuda a visualizar a relação entre a localização e o valor das casas.
```

#Célula 9: Subamostragem para Visualização


```python
# Subamostragem para visualização
rng = np.random.RandomState(0)
indices = rng.choice(np.arange(california_housing.frame.shape[0]), size=500, replace=False)
sns.scatterplot(
    data=california_housing.frame.iloc[indices],
    x="Longitude",
    y="Latitude",
    size="MedHouseVal",
    hue="MedHouseVal",
    palette="viridis",
    alpha=0.5,
)
plt.legend(title="MedHouseVal", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.title("Median house value depending of their spatial location (Subsampled)")
plt.show()

# Subamostragem para melhorar a visualização
# - rng.choice: Seleciona aleatoriamente 500 pontos do dataset para reduzir a sobrecarga de dados no gráfico.
# - Isso mantém a clareza dos padrões geográficos sem comprometer a visualização.
```

#Célula 10: Pair Plot das Features

```python
# Pair plot das features (excluindo longitude e latitude)
columns_drop = ["Longitude", "Latitude"]
subset = california_housing.frame.iloc[indices].drop(columns=columns_drop)
subset["MedHouseVal"] = pd.qcut(subset["MedHouseVal"], 6, retbins=False)
subset["MedHouseVal"] = subset["MedHouseVal"].apply(lambda x: x.mid)
sns.pairplot(data=subset, hue="MedHouseVal", palette="viridis")
plt.show()

# Criando um pair plot das features (excluindo longitude e latitude)
# - pairplot: Mostra as relações entre todas as features em um único gráfico.
# - pd.qcut: Divide o target (MedHouseVal) em 6 intervalos para facilitar a visualização.
# - Isso ajuda a identificar correlações e padrões entre as features.
```

#Célula 11: Criação e Avaliação do Modelo de Regressão Ridge

```python
# Criando e avaliando um modelo de regressão Ridge
alphas = np.logspace(-3, 1, num=30)  # Intervalo de valores para o parâmetro alpha
model = make_pipeline(StandardScaler(), RidgeCV(alphas=alphas))  # Pipeline com normalização e RidgeCV

# Validação cruzada
cv_results = cross_validate(
    model,
    california_housing.data,  # Features
    california_housing.target,  # Target
    return_estimator=True,      # Retorna os modelos treinados em cada fold
    n_jobs=2,                   # Usa 2 núcleos do processador
)

# Exibindo o score R² médio
score = cv_results["test_score"]
print(f"R2 score: {score.mean():.3f} ± {score.std():.3f}")



# Criando e avaliando um modelo de regressão Ridge
# - alphas: Define um intervalo de valores para o parâmetro de regularização alpha.
# - make_pipeline: Cria um pipeline que primeiro normaliza os dados (StandardScaler) e depois aplica o modelo RidgeCV.
# - cross_validate: Realiza validação cruzada para avaliar o desempenho do modelo.
# - O score R² é exibido para indicar o quão bem o modelo está performando.
```

```python
# Visualizando os coeficientes do modelo Ridge
coefs = pd.DataFrame(
    [est[-1].coef_ for est in cv_results["estimator"]],  # Extrai os coeficientes de cada modelo
    columns=california_housing.feature_names,            # Nomes das features
)

# Definindo cores para o gráfico de caixa (box plot)
color = {"whiskers": "black", "medians": "black", "caps": "black"}

# Plotando o gráfico de caixa dos coeficientes
coefs.plot.box(vert=False, color=color)

# Adicionando uma linha vertical em x=0 para referência
plt.axvline(x=0, ymin=-1, ymax=1, color="black", linestyle="--")

# Adicionando título ao gráfico
plt.title("Coefficients of Ridge models via cross-validation")

# Exibindo o gráfico
plt.show()


# Visualizando os coeficientes do modelo Ridge
# - coefs: Cria um DataFrame com os coeficientes do modelo para cada fold da validação cruzada.
# - coefs.plot.box: Cria um gráfico de caixa (box plot) para visualizar a distribuição dos coeficientes.
# - plt.axvline: Adiciona uma linha vertical em x=0 para referência.
# - Isso ajuda a entender a importância e a consistência de cada feature no modelo.
```

```python
# Importando bibliotecas necessárias
from sklearn.datasets import fetch_california_housing
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import RidgeCV
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_validate
import joblib
from google.colab import drive

# Montar o Google Drive
drive.mount('/content/drive')

# Carregar o dataset
california_housing = fetch_california_housing(as_frame=True)

# Criar e avaliar um modelo de regressão Ridge
alphas = np.logspace(-3, 1, num=30)  # Intervalo de valores para o parâmetro alpha
model = make_pipeline(StandardScaler(), RidgeCV(alphas=alphas))  # Pipeline com normalização e RidgeCV

# Treinar o modelo com todos os dados
model.fit(california_housing.data, california_housing.target)

# Salvar o modelo no Google Drive
model_path = "/content/drive/My Drive/california_housing_model2.pkl"
joblib.dump(model, model_path)

print(f"Modelo salvo em: {model_path}")
```
